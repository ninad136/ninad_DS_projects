### Hypothesis Testing :

```python
# 
"""central limit theorem states that the distribution of 
the sample is normal if the sample size is large or more than 30."""
```

### Wilcoxon Rank-sum test for  continuity hypothesis testing :

```python
from scipy.stats import ranksums

df1=df["went_on_backorder"]
df2=df["national_inv"]==0
ranksums(df1,df2)
```

```python
# the p value is less than the alpha value 0.05 we can reject the null hypothesis
```

```python
df1=df["went_on_backorder"]
df2=df["sales_9_month"]
ranksums(df1,df2)
```

```python
df1=df["went_on_backorder"]
df2=df["sales_9_month"]==df["sales_9_month"].max()
ranksums(df1,df2)
```

```python
df1=df["went_on_backorder"]
df2=df["sales_9_month"]==df["sales_9_month"].min()
ranksums(df1,df2)
```

```python
# feature chosen as per the heatmap
features=["national_inv","lead_time","sales_1_month","sales_3_month","sales_6_month",
          "sales_9_month","forecast_3_month","forecast_6_month","forecast_9_month"]
```

```python
X=df[features]
```

```python
X.info()
```

```python
y=df["went_on_backorder"]
```

```python
X.head()
```

```python
# scaling the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
X=pd.DataFrame(X,columns=features)
```

```python
X.head()
```

```python
# splitting into training and testing 
# split the data in train test evaluate
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,stratify=y)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
# model building
```

```python
from sklearn.metrics import roc_auc_score,confusion_matrix,accuracy_score,classification_report
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

### Logistic Regression :

```python
# logistic Regression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# predict the model
Y_pred_lr=lr.predict(X_test)
```

```python
# model performance
print("Logistic regression performance \n",lr.score(X_train,y_train))
print("LR \n",accuracy_score(y_test,Y_pred_lr))
print("confusion matrix \n",confusion_matrix(y_test,Y_pred_lr))
print("classification report \n",classification_report(y_test,Y_pred_lr))
```

### Decision Tree :

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
# predict the model
Y_pred_dt=dt.predict(X_test)
```

```python
# model performance
print("DT performance \n",dt.score(X_train,y_train))
print("DT \n",accuracy_score(y_test,Y_pred_dt))
print("confusion matrix \n",confusion_matrix(y_test,Y_pred_dt))
print("classification report \n",classification_report(y_test,Y_pred_dt))
```

### Random Forest Classifier :

```python
#RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=15,min_samples_leaf=5,oob_score=True)
rf.fit(X_train,y_train)
```

```python
# predict the model
Y_pred_rf=rf.predict(X_test)
```

```python
# model performance
print("RF performance \n",rf.score(X_train,y_train))
print("RF \n",accuracy_score(y_test,Y_pred_rf))
print("confusion matrix \n",confusion_matrix(y_test,Y_pred_rf))
print("classification report \n",classification_report(y_test,Y_pred_rf))
```
